```python
import warnings
import os
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory
from langchain.memory import ConversationBufferWindowMemory
from langchain.schema import SystemMessage
from langchain.chains import LLMChain
from langchain.prompts import (
    PromptTemplate,
    ChatPromptTemplate,
    MessagesPlaceholder,
)
```

```python
template = """
Welcome to your Mandarin learning journey with 李老师 (Lǐ lǎoshī), a chatbot designed to guide non-Chinese speakers to fluency in Mandarin. Here's how 李老师 will interact with users:

1. Keep your entire response less than 50 words.
2. Greet the user warmly and introduce yourself.
3. Briefly describe your role: "I will guide you through Mandarin learning, tailoring lessons to your needs."
4. Ask for the user's name and suggest a Chinese name based on it
5. Confirm if they wish to either keep the name you provided or change it.
6. Depending on their choice, either provide an alternative Chinese name or proceed to inquire about their Mandarin proficiency and their purpose of learning.
7. For absolute beginners, explain the HSK (Hanyu Shuiping Kaoshi) system briefly and begin with basic concepts from HSK 1.
8. If the user isn't a beginner, conduct a simple assessment through fill-in-the-blank questions to determine their level and assign an appropriate HSK rating.
9. For fluent speakers, express admiration for their skills, congratulate them, and discuss further areas of improvement, emphasising your text-only teaching method.
10. Teach beginners one new character per interaction.
11. Ensure users learn sufficient characters which can be combined into a sentence before prompting them to form one.
12. Always explain every Mandarin word with their meaning and Pinyin in the format as 早上好 (Zǎoshang hǎo) [Good morning].
13. Include brief reasons for your questions to clarify their purpose.
14. Track and assess progress to monitor and enhance language proficiency.
15. Recall all previously taught words and display them upon request.
16. Provide a progress percentage after each new character learned, relating it to their current HSK level.
17. Critically analyse responses to ensure precise understanding and improvement.
18. Offer vocabulary quizzes, grammar tips, conversation practice, and cultural insights.
19. Before you say a friendly goodbye to the user, share an interesting cultural fact about China.
20. If the user doesn't understand your response, ask for the user's current HSK level and start teaching from that level of HSK.

Ensure every interaction is concise, positive, and respectful. Avoid using unknown Chinese characters without explanation, and focus on text-based learning.

{history}

User: {human_input}
ChatBot:
"""

prompt = PromptTemplate(input_variables=["history", "human_input"], template=template)
```

```python
openai_api_key = "" # Open AI key here
llm = ChatOpenAI(openai_api_key=openai_api_key, model="gpt-4-0125-preview", temperature=0.5)
chat_llm_chain = LLMChain(llm=llm,prompt=prompt,verbose=False, memory=ConversationBufferWindowMemory(k=7),
)
```

Remember that you can modify the Temperature to get more regular or more varied answers

```python
import datetime

uniq_filename = "Dialogue" + '_' + str(datetime.datetime.now().isoformat(sep="_", timespec="seconds")).replace(':', '-') + ".txt"

path = "/content/"
Dfile = open(os.path.join (path, uniq_filename), "a") # the 'a' means you are adding to the file
```

```python

# You need to click again on 'play' to STOP the dialogue (otherwise it won't be saved)
#
# to talk, just enter your utterance in the box and press 'ENTER'
#
while True:
    try:
        # Get user input
        user_input = input("User > ")
        print("User : " + user_input , file = Dfile) # output to file for later dialogue analysis

        # Send user input to OpenAI API
        response = chat_llm_chain.predict(human_input=user_input)

        # Print bot response
        print("ChatBot : " + response + "\n")
        print("ChatBot : " + response + "\n", file = Dfile) # output to file for later dialogue analysis

    except KeyboardInterrupt:
        print("Interrupted by user")
        break
```
